##Prepara√ß√£o dos Dados:

Carregar bibliotecas e base de dados iris

```python
#Carregar bibliotecas python
import random
random.seed(7) # defini uma semente (importante para reproduzir resultados)
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.cluster import KMeans

from sklearn.model_selection import train_test_split  #particoes treino e teste
from sklearn.metrics import accuracy_score  #verifica a exatid√£o do modelo
```

```python
#Carregar os dados
##########BAIXAR O DATASET E DEPOIS FAZER O UPLOAD .csv NO GOOGLE COLAB#########
#fonte dos dados: https://www.kaggle.com/datasets/uciml/iris
################################################################################

#Carregar os dados num dataframe
data = pd.read_csv('/content/Iris.csv', header=(0))


#armazena os nomes das classes
classes = np.array(pd.unique(data[data.columns[-1]]), dtype=str)

print("Confira o n√∫mero de linhas e colunas na matriz de atributos:", data.shape)
attributes = list(data.columns)
data.head(10)  #ver os 10 primeiros exemplares da base de dados
```

```python
# Verifica o nome das classes
print(classes)
```

Parti√ß√µes: Selecionando os conjuntos de treinamento e teste.

```python
#Preprando os dados
data = data.to_numpy()  #transforma o dataframe em um vetor numpy
nrow, ncol = data.shape  #ler o n√∫mero de linhas e colunas da base de dados
y = data[:, -1]  #armazena informa√ß√µes da vari√°vel resposta (classe)
X = data[:,0:ncol-1]  #separa os dados num√©ricos da classe
```

```python
X
```

```python
y
```

```python
#Verificar as dismens√µes do vetor X e y
print(X.shape, y.shape)
```

```python
#constroi a particao de treino e teste
particao = 0.7 # fracao de elementos no conjunto de treinamento
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    train_size = particao,
                                                    random_state = 7)
```

##Classifica√ß√£o

Usando a biblioteca scikit-learn

###Por algoritmo Naive-Bayes (Aprendizado Supervisionado)

```python
#carrega bibliotecas com Naive Bayes implementado
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics

model = GaussianNB()  #instancia o objeto para receber X_train e y_train
model.fit(X_train, y_train)  #treina o modelo com o mesmo X_train e y_train

y_pred_sktlearn = model.predict(X_test)  #realiza a predi√ß√£o da particao teste
score = accuracy_score(y_pred_sktlearn, y_test)  #calcula a accuracy do modelo
print('Accuracy:', score)
```

###K-means  (Aprendizado N√£o Supervisionado)

```python
#Inicializa√ß√£o dos par√¢mtros K-means
kmeans_kwargs = {"init": "random", "n_init": 10,"max_iter": 500, "random_state": 42}
kmeans = KMeans(n_clusters=3, **kmeans_kwargs)
```

```python
#Alimentar a instancia do Kmeans com os dados
kmeans.fit(X)
```

```python
#Preparar as classifica√ß√µes do algoritmo para avaliar resultados
kmeans.labels_.reshape(-1,1)
```

```python
#Associar as classifica√ß√µes K-means com os r√≥tulos original dos dados
np.concatenate((kmeans.labels_.reshape(-1,1), y.reshape(-1,1)), axis= 1)
```

```python
from scipy.stats.contingency import crosstab
```

```python
pd.crosstab(kmeans.labels_, y)
```

Diante dos resultados do quadro acima, podemos fazer as seguintes associa√ß√µes:

- A coluna setosa est√° associada ao clsuter 1;
- A coluna versicolor est√° associada ao clsuter 0, pois temos 48 classifica√ß√µes favor√°veis a este cluster em rela√ß√£o ao cluster 2; e
- A virginica est√° associada ao clsuter 2, pois h√° majoritariamente 36 classfica√ß√µes vs. 14 associadas ao cluster 0, que inclusive j√° estava associado ao r√≥tulo versicolor.

Desta forma, temos a indecis√£o de 16 observa√ß√µes, que ser√£o classificadas erroneamente ao atribuirmos o cluster ao r√≥tulo majorit√°rio.

Logo, temos um n√≠vel de accuracy de $\frac{150 - 14 - 2}{150} = \frac{134}{150}$, que √©:

```python
#C√°lculo da accuracy
134/150
```

Uma vez que o k-means sugeriu os r√≥tulos e usando um tabela contig√™ncia acima, podemos fazer o uso de uma matriz de confus√£o j√° implemetada no scikit-learn, conforme c√≥digo a seguir.

```python
#Colocar dados no dataframe
df = pd.DataFrame(np.concatenate((kmeans.labels_.reshape(-1,1), y.reshape(-1,1)), axis= 1), columns=['kmeans', 'original'])
```

```python
df
```

```python
#Substituir valores de acordo com resultados da tabela de contig√™ncia
df.replace(1, 'versicolor', inplace = True )
df.replace(0, 'virginica', inplace = True )
df.replace(2, 'setosa', inplace = True )
```

```python
df
```

```python
df.kmeans.unique()
```

```python
#Usar uma matriz de confusao para avaliar a accuracy
score = accuracy_score(df.kmeans, y)  #calcula a accuracy do modelo
print('Accuracy:', score)
```
